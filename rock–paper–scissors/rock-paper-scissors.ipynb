```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list

from random import choice

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.

outcomes = ['stone', 'paper', 'scissors']

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...

numOfGames = 5

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games

mustWin = int(numOfGames / 2) + 1

print("Total games: {}".format(numOfGames),
      "\nMust win {} out of {}".format(mustWin, numOfGames))

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.

def machineChoice(list = outcomes):
    return choice(list)


# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.

def userChoice(list = outcomes):
    user_choice = input("\nMake your choice:\n{0}, {1}, {2}: ".format(
        list[0], list[1], list[2])).casefold() # coverts input to lowercase
    while (user_choice not in list):
        print("\nPlease make sure your choice is spelled correctly\n")
        user_choice = input("Make your choice:\n{0}, {1}, {2}: ".format(
            list[0], list[1], list[2])).casefold() # prompts until proper input provided
    else:
        return user_choice
   


# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins

# dictionary containing win scenarios

beats = {
    'scissors' : 'stone', # 'stone beats scissors'
    'paper' : 'scissors', # 'scissors beats paper'
    'stone' : 'paper'     # 'paper beats rock'
}

def combat():
    if (user == machine): # returns 0 for a tie
        combat.result = 0
        return 0
    elif (machine == beats[user]): # machine wins
        combat.result = 1
        return 1
    elif (user == beats[machine]): # user wins
        combat.result = 2
        return 2
        
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated

def gameState():
    print("\n------Game {}------\n".format(gameCount))
    if result == 0: 
        print("\nYou:     {0}\nMachine: {1}\n\nTie\n".format(user, machine))
    elif result == 1:
        print("\nYou:     {0}\nMachine: {1}\n\nMachine wins\n".format(user, machine))
    elif result == 2:
        print("\nYou:     {0}\nMachine: {1}\n\nYou win\n".format(user, machine))
    else:
        pass

    
# Create two variables that accumulate the wins of each participant

userWins = 0
machineWins = 0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.

gameCount = 1

while (userWins < mustWin) and (machineWins < mustWin) and (gameCount <= numOfGames):
    user = userChoice()
    machine = machineChoice()
    result = combat()
    gameState()
    if (result == 1):
        machineWins += 1
    elif (result == 2):
        userWins += 1
    else:
        pass # in case of a 'Tie', continue with loop
    print("\n---Current Score---\n")
    print("Machine: {}".format(machineWins), "\tYou: {}\n".format(userWins))
    gameCount += 1

    
# Print by console the winner of the game based on who has more accumulated wins
  
if (userWins >= mustWin):
    print("\nYou won the game!")
elif (machineWins >= mustWin):
    print('\n"You\'ve been defeated" \n- Machine :)')
else:
    print("\nNeither side has scored enough to win. Try again.")
```
